```python
'''Problem 6 : Sum of Digits
Write a recursive function to calculate the sum of digits of a given positive integer.
sum_of_digits(123) -> 6
'''
def sum_of_digits(n):
    # Base case: if the number is a single digit
    if n < 10:
        return n
    # Recursive case: sum the last digit and call the function with the remaining digits
    else:
        return n % 10 + sum_of_digits(n // 10)

# Example usage:
number = 123
result = sum_of_digits(number)

print(f"The sum of digits of {number} is: {result}")
```

```python
'''Problem 7: Fibonacci Series
Write a recursive function to generate the first n numbers of the Fibonacci series.
fibonacci_series(6) -> [0, 1, 1, 2, 3, 5]'''
def fibonacci_series(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fib_series = fibonacci_series(n - 1)
        fib_series.append(fib_series[-1] + fib_series[-2])
        return fib_series

# Example usage:
n = 6
result = fibonacci_series(n)

print(f"The first {n} numbers of the Fibonacci series are: {result}")
```

```python
'''
Problem 8 : Subset Sum
Given a set of positive integers and a target sum, write a recursive function to determine if there exists a subset
of the integers that adds up to the target sum.
subset_sum([3, 34, 4, 12, 5, 2], 9) -> True'''
def subset_sum(nums, target_sum):
    # Base cases
    if target_sum == 0:
        return True
    if not nums or target_sum < 0:
        return False

    # Recursive cases: check with and without the last element
    return subset_sum(nums[:-1], target_sum - nums[-1]) or subset_sum(nums[:-1], target_sum)

# Example usage:
numbers = [3, 34, 4, 12, 5, 2]
target = 9
result = subset_sum(numbers, target)

print(f"There exists a subset that adds up to {target}: {result}")
```

```python
'''
Problem 9: Word Break
Given a non-empty string and a dictionary of words, write a recursive function to determine if the string can be
segmented into a space-separated sequence of dictionary words.
word_break( leetcode , [ leet , code ]) -> True'''
def word_break(s, word_dict):
    # Base case: an empty string is always breakable
    if not s:
        return True

    # Check all possible prefixes of the current string
    for prefix in word_dict:
        if s.startswith(prefix):
            # Recursively check the remaining substring after removing the prefix
            if word_break(s[len(prefix):], word_dict):
                return True

    # If no valid break is found, return False
    return False

# Example usage:
string_to_break = "leetcode"
dictionary = ["leet", "code"]
result = word_break(string_to_break, dictionary)

print(f"The string '{string_to_break}' can be segmented: {result}")
```

```python
'''Problem 10 : N-Queens
Implement a recursive function to solve the N Queens problem, where you have to place N queens on an NÃ—N
chessboard in such a way that no two queens threaten each other.
n_queens(4)
[
[".Q..",
"...Q",
"Q...",
"..Q."],
["..Q.",
"Q...",
"...Q",
".Q.."]
]'''
def is_safe(board, row, col, n):
    # Check if there is a queen in the same row
    for i in range(col):
        if board[row][i] == 'Q':
            return False

    # Check if there is a queen in the upper diagonal
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 'Q':
            return False

    # Check if there is a queen in the lower diagonal
    for i, j in zip(range(row, n), range(col, -1, -1)):
        if board[i][j] == 'Q':
            return False

    return True

def solve_n_queens_util(board, col, n, result):
    if col == n:
        result.append(["".join(row) for row in board])
        return

    for i in range(n):
        if is_safe(board, i, col, n):
            board[i][col] = 'Q'
            solve_n_queens_util(board, col + 1, n, result)
            board[i][col] = '.'

def solve_n_queens(n):
    result = []
    board = [['.' for _ in range(n)] for _ in range(n)]
    solve_n_queens_util(board, 0, n, result)
    return result

# Example usage:
n = 4
queens_arrangements = solve_n_queens(n)
print(queens_arrangements)
```
